```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./me_climbing_deaths.csv')
```

```python
df.head()
```

```python
df.describe(include='all')
```

```python
# Age kolonnen mangler 150 values som er null, så man kunne principelt overveje at fjerne denne kolonne.
# 48.5%

df.info()
```

```python
# Her har vi filled missing data fra "Age" kolonnen med mode.
df['Age'].fillna(value=df['Age'].mode()[0], inplace=True)
```

```python
# Her har vi filled missing data fra "Nationality" kolonnen med mode.
df['Nationality'].fillna(value=df['Nationality'].mode()[0], inplace=True)

nth_row = df.iloc[275]

nth_row
```

```python
# Her har vi filled missing data fra "Expedition" kolonnen med mode.
# Ændrer Expedition til "Solo", da det generelt ikke er samme datoer, samtidigt med at det er udelukket at solo climb mt. Everest.
df['Expedition'].fillna(value='Solo', inplace=True)

df.info()
```

```python
# Her har vi filled missing data fra "Cause of death" kolonnen med mode.
df['Cause of death'].fillna(value='Unknown cause', inplace=True)

df.info()
```

```python
# Her har vi filled missing data fra "Location" kolonnen med mode.
df['Location'].fillna(value=df['Location'].mode()[0], inplace=True)

df.info()
```

```python


# Check the completeness of the 'Age' column
age_data_complete = df['Age'].dropna()

# Plotting for Age
plt.figure(figsize=(15, 5))

# Check if there are any non-null age data
if len(age_data_complete) > 0:
    # Histogram for Age
    plt.subplot(1, 2, 1)
    sns.histplot(age_data_complete, kde=True)
    plt.title('Age Distribution')
    plt.xlabel('Age')
    plt.ylabel('Frequency')

    # Boxplot for Age
    plt.subplot(1, 2, 2)
    sns.boxplot(x=age_data_complete)
    plt.title('Age Boxplot')
    plt.xlabel('Age')
else:
    print("Insufficient age data for visualization.")

plt.tight_layout()
plt.show()

# Setting up the figure for three bar charts
plt.figure(figsize=(18, 6))

# Bar chart for Nationality
plt.subplot(1, 3, 1)
nationality_counts = df['Nationality'].value_counts()
sns.barplot(x=nationality_counts[:10], y=nationality_counts.index[:10])
plt.title('Top 10 Nationalities')
plt.xlabel('Count')
plt.ylabel('Nationality')

# Bar chart for Cause of Death
plt.subplot(1, 3, 2)
cause_of_death_counts = df['Cause of death'].value_counts()
sns.barplot(x=cause_of_death_counts[:10], y=cause_of_death_counts.index[:10])
plt.title('Top 10 Causes of Death')
plt.xlabel('Count')
plt.ylabel('Cause of Death')

# Bar chart for Location
plt.subplot(1, 3, 3)
location_counts = df['Location'].value_counts()
sns.barplot(x=location_counts[:10], y=location_counts.index[:10])
plt.title('Top 10 Locations of Death')
plt.xlabel('Count')
plt.ylabel('Location')

plt.tight_layout()
plt.show()
```
