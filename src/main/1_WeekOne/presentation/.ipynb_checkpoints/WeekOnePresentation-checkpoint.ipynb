Below is an example where we use %time to calculate the time it takes to run one code cell. It shows the user time, system time and the total time.

```python
%time squares = [x**2 for x in range(1, 100)]
```

```python
%%time
x = [n**2 for n in range(100)]
y = [n**3 for n in range(100)]
```

# Flow of control in Python

The order in which the program's code executes.
Using:
- conditionlan statements: if, elif, else
- Loops: for, while
- function calls.

```python
for i in range(1, 11):
    if i % 2 == 0:
        continue
    print(i)
```

```python
def is_even(number):
    return number % 2 == 0

def print_numbers_and_status(numbers):
    for number in numbers:
        if is_even(number):
            print(f"{number} is even")
        else:
            print(f"{number} is odd")

numbers = range(1, 6)  
print_numbers_and_status(numbers)
```

```python
def greeting(name):
    print(f"Hello, {name}!")

while True:
    name = input("Enter your name (or 'quit' to quit): ")
    if name.lower() == 'quit':
        break
    else:
        greeting(name)
```

```python
x = 10

if x > 5:
    print("x is greater than 5")
else:
    print("x is not greater than 5")
```

# Data types & data structures

```python
Data Types:

Text Type:	str
Numeric Types:	int, float, complex
Sequence Types:	list, tuple, range
Mapping Type:	dict
Set Types:	set, frozenset
Boolean Type:	bool
Binary Types:	bytes, bytearray, memoryview
None Type:	NoneType
```

```python
# Integer
x = 5

# Float
y = 3.14

# String
name = "Alice"

# Boolean
is_student = True

print(x,type(x))               
print(y,type(y))                 
print(name,type(name))              
print(is_student,type(is_student))       
```

```python
colors = ["red", "green", "blue"]
print(colors,type(colors))
```

```python
dict = {"firstName": "Mark", "lastName": "Lundgaard"}
print(dict)

print("Printing by using key:")
print(dict["firstName"])

print("Printing by using get:")
print(dict.get("firstName"))

print("Printing by using Dictionary Comprehension")
myDict = {x: x**2 for x in [1,2,3,4,5]}
print(myDict,type(myDict))
```

```python
my_tuple = (1,2,3)
print(my_tuple,type(my_tuple))
```

# Numpy

```python
# numpy

import numpy as np

# Creating a 1D array with 12 elements
original_array = np.arange(1, 13)  # This will create an array with elements from 1 to 12

print("Original Array:")
print(original_array)

# Reshaping the array to 3 rows and 4 columns
reshaped_array = np.reshape(original_array, (3, 4))

print("\nReshaped Array:")
print(reshaped_array)
```

# Matplotlib

```python
# %matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x))
plt.show()
```

```python
# matplotlib

import matplotlib.pyplot as plt

# Sample data
x = [0, 1, 2, 3, 4, 5]  # x-axis values
y = [0, 1, 4, 9, 16, 25]  # y-axis values (y = x^2)

# Creating the plot
plt.plot(x, y)

# Adding title
plt.title('Sample Plot')

# Adding labels to the axes
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')

# Showing the plot
plt.show()
```

# Seaborn

```python
# seaborn

import seaborn as sns
import matplotlib.pyplot as plt

# Loading an example dataset from Seaborn called tips
tips = sns.load_dataset("tips")

# Creating a scatter plot
# Here we plot 'total_bill' on the x-axis and 'tip' on the y-axis
sns.scatterplot(x="total_bill", y="tip", data=tips)

# Adding a title to the plot
plt.title('Tip Amount vs Total Bill')

# Showing the plot
plt.show()

# This scatter plot shows the relationship between the total bill and tip amount.
```
